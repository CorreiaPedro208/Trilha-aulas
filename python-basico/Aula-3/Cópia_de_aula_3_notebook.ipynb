## Questão 1 📝

Crie um programa que pede dois números ao usuário e exibe o resultado da divisão.

Trate dois erros possíveis de forma **separada**:
- O usuário digitar algo que não seja número (`ValueError`)
- O usuário digitar zero no denominador (`ZeroDivisionError`)

Use também o bloco `else` para exibir o resultado apenas quando tudo funcionar, e o bloco `finally` para sempre imprimir `"Operação encerrada."`

````python
try:
  dividendo = float(input("Digite o dividendo:"))
  divisor = float(input("Digite o divisor:"))
  divisao = dividendo/divisor

except ValueError as error:
  print(f"Erro: {error}")
except ZeroDivisionError as Error:
  print(f"Erro: {Error}")
else:
  print(f"O valor da divisão é: {divisao}")
finally:
  print("Operação encerrada.")
````

### 💡 Dica:

*Utilize somente se necessário — tente resolver primeiro sozinho!*

| Bloco | Quando executa |
|---|---|
| `except ValueError` | entrada inválida (letras, símbolos) |
| `except ZeroDivisionError` | divisão por zero |
| `else` | **apenas** se nenhum erro ocorreu no `try` |
| `finally` | **sempre**, com ou sem erro |

## Questão 2 📝

Crie uma função `calcular_raiz(numero)` que:
- Lança um `ValueError` com a mensagem `"Número não pode ser negativo."` se o argumento for negativo
- Retorna a raiz quadrada do número caso contrário

Chame a função dentro de um `try/except` que capture o erro e imprima a mensagem.

*Teste chamando a função com `-9`.*

````python
def calcular_raiz(numero):
  if numero <0:
    raise ValueError("Número não pode ser negativo")
  return numero**0,5

try:
  resultado = calcular_raiz(-9)
  print(f"A raiz é: {resultado}")
except ValueError as Error:
  print(f"Erro:{Error}")


````

### 💡 Dica:

*Utilize somente se necessário — tente resolver primeiro sozinho!*

`raise` interrompe a função e lança o erro que você escolher.

O `as erro` no `except` captura o objeto do erro — com ele você consegue imprimir a mensagem que foi passada ao `ValueError`.

## Questão 3 📝

Crie um programa que:
1. Abre (ou cria) um arquivo chamado `"notas.txt"` no modo escrita
2. Pede ao usuário o nome e a nota de 3 alunos
3. Salva cada entrada no arquivo no formato: `"Nome: nota"`
4. Ao final, imprime `"Notas salvas com sucesso!"`

````python
with open("notas.txt", "w", encoding="utf-8") as arquivo:
  for i in range(3):
    nome = input("Digite o nome do aluno: ")
    nota = float(input("Digite a nota do aluno: "))
    arquivo.write(f"{nome}: {nota}\n")
print("Notas salvas com sucesso!")
with open("notas.txt", "r", encoding="utf-8") as arquivo:
  print(arquivo.read())



````

### 💡 Dica:

*Utilize somente se necessário — tente resolver primeiro sozinho!*

Use `write()` para escrever no arquivo. Diferente do `print()`, ele **não** adiciona quebra de linha automaticamente — coloque `\n` ao final da string.

Sempre use `encoding="utf-8"` para evitar problemas com acentos.

## Questão 5 📝

Use a biblioteca padrão `random` para criar um simulador de dado de 6 faces.

O programa deve:
1. Sortear um número inteiro entre 1 e 6
2. Imprimir o resultado no formato: `"Você tirou: X"`
3. Depois, criar uma lista com 5 frutas e sortear uma aleatoriamente

````python
import random
resultado = random.randint(1,6)
print(f"Você tirou: {resultado}")

lista = ["Banana", "Maçã", "Melão", "Jabuticaba", "Abacate"]
fruta = random.choice(lista)
print(f"A fruta é: {fruta}")
````

### 💡 Dica:

*Utilize somente se necessário — tente resolver primeiro sozinho!*

Bibliotecas padrão já vêm no Python — basta usar `import`.

- `random.randint(a, b)` → inteiro aleatório entre `a` e `b` (**incluindo os dois**)
- `random.choice(lista)` → elemento aleatório de uma lista
